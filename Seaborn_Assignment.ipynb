Q1.Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

1.Barplot():

A barplot is basically used to aggregate the categorical data according to some methods and by default itâ€™s the mean. It can also be understood as a visualization of the group by action. To use this plot we choose a categorical column for the x-axis and a numerical column for the y-axis, and we see that it creates a plot taking a mean per categorical column.

```python
import seaborn as sns
tip=sns.load_dataset("tips")
sns.barplot(x="day",y="tip",data=tip)
```

2.scatterplot():

Seaborn scatterplot is a visualization tool used to explore the relationship between two variables. It is a type of plot that displays the values of two variables as points on a two-dimensional coordinate system.
The primary use of a seaborn scatterplot is to identify patterns or trends in the data. It helps in understanding the relationship between two variables, such as whether they are positively or negatively correlated.
In addition to exploring relationships between two variables, seaborn scatterplots can also be used to compare multiple groups of data. By adding color or size to the points on the scatterplot, different groups of data can be distinguished and compared.

```python
sns.scatterplot(x="total_bill",y="tip",data=tip)
```

3.displot(): 

The primary use of seaborn displot() is to visualize the distribution of a dataset and to identify any underlying patterns or trends. It helps to explore the shape of the data distribution, such as whether it is symmetric or skewed, and to identify any outliers or unusual values.

```python
sns.displot(tip["total_bill"])
```

4.pairplot(): 

Pairplot displays scatterplots between all possible pairs of variables in a dataset, and can also include histograms or kernel density plots of the variables along the diagonal of the plot. This allows for a quick overview of the relationships between variables in a dataset.
The primary use of seaborn pairplot is to identify patterns or trends between pairs of variables in a dataset. It can be used to explore the correlation between variables, to identify potential outliers or data errors, and to visualize the distribution of the data.

```python
sns.pairplot(tip)
```

5.Stripplot():
    
It basically creates a scatter plot based on the category.

```python
sns.stripplot(x="day",y="total_bill",data=tip,hue="sex")
```
